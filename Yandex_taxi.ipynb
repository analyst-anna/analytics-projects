```python
df = pd.read_csv('yandex_taxi.csv', sep=';')
```

```python
print(f'Столбцов: {len(df.columns)}')
print(df.columns.tolist())
```

Проводим очистку данных

Ошибка вызвана несуществующей датой 2024-11-31 (в ноябре 30 дней).
Обработаем дату с errors='coerce' для замены некорректных значений на NaT. Удалим строки с некорректными данными

```python
df['Date and time of start of the trip'] = pd.to_datetime(df['Date and time of start of the trip'], format='%d.%m.%Y %H:%M', errors='coerce')
```

```python
print(df[df['Date and time of start of the trip'].isna()])
```

```python
df = df.dropna(subset=['Date and time of start of the trip'])
```

 Анализ и обработка пропусков

```python
# 1. Проверим пропуски
print("Пропуски по столбцам ДО очистки:")
print(df.isnull().sum())

# 2. Удалим строки с пропусками в ключевых столбцах
key_columns = ['start_lat', 'start_lon', 'end_lat', 'end_lon', 'Cost', 'Length of the trip']
df = df.dropna(subset=key_columns)

# 3. Проверим результат
print("\nПропуски по столбцам ПОСЛЕ очистки:")
print(df.isnull().sum())
```

Работа с категориальными признаками

```python
# 1. Приведем к нижнему регистру и уберем пробелы
categorical_cols = ['Weather', 'Time in a day', 'Day of the week']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.lower()

# 2. Проверим уникальные значения
print("\nУникальные значения:")
for col in categorical_cols:
    if col in df.columns:
        print(f"\n{col}: {df[col].unique()}")
```

Верификация координат

```python
# Установим разумные границы для Санкт-Петербурга
LAT_MIN, LAT_MAX = 59.7, 60.2
LON_MIN, LON_MAX = 30.1, 30.6

# Посчитаем строки с выбросами ДО фильтрации
print(f"Строк до фильтрации координат: {len(df)}")

# Отфильтруем выбросы
df = df[
    (df['start_lat'].between(LAT_MIN, LAT_MAX)) &
    (df['start_lon'].between(LON_MIN, LON_MAX)) &
    (df['end_lat'].between(LAT_MIN, LAT_MAX)) &
    (df['end_lon'].between(LON_MIN, LON_MAX))
]

print(f"Строк после фильтрации координат: {len(df)}")
```

Проверка логики данных и аномалий

```python
# 1. Проверим корреляцию длины и стоимости
correlation = df['Length of the trip'].corr(df['Cost'])
print(f"\nКорреляция между длиной и стоимостью: {correlation:.3f}")

# 2. Удалим очевидные аномалии
df = df[
    (df['Length of the trip'] > 0.1) &  # Минимальная разумная длина
    (df['Length of the trip'] < 100) &  # Максимальная разумная длина
    (df['Cost'] > 50) &                 # Минимальная разумная стоимость
    (df['Cost'] < 10000)                # Максимальная разумная стоимость
]

print(f"Строк после удаления аномалий: {len(df)}")
```

Создание новых признаков

```python
# 1. Извлечем временные признаки из даты
df['hour'] = df['Date and time of start of the trip'].dt.hour
df['day_of_week_num'] = df['Date and time of start of the trip'].dt.dayofweek
df['month'] = df['Date and time of start of the trip'].dt.month

# 2. Создадим признаки времени суток
def time_of_day_period(hour):
    if 5 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 17:
        return 'afternoon'
    elif 17 <= hour < 23:
        return 'evening'
    else:
        return 'night'

df['time_period'] = df['hour'].apply(time_of_day_period)

# 3. Проверим результат
print("\nСозданные признаки:")
print(df[['hour', 'day_of_week_num', 'month', 'time_period']].head())
```

Делаем разворот таблицы для создания полилиний

```python
# Создаем уникальный ID для каждой поездки
df['id_trip'] = df.index + 1

# Создаем таблицу для стартовых точек
df_start = df[['id_trip', 'start_lat', 'start_lon', 'Date and time of start of the trip']].copy()
df_start = df_start.rename(columns={
    'start_lat': 'lat',
    'start_lon': 'lon',
    'Date and time of start of the trip': 'timestamp'
})
df_start['point_type'] = 'start'
df_start['point_order'] = 1  # Порядок точек в линии

# Создаем таблицу для конечных точек
df_end = df[['id_trip', 'end_lat', 'end_lon', 'Date and time of start of the trip']].copy()
df_end = df_end.rename(columns={
    'end_lat': 'lat',
    'end_lon': 'lon',
    'Date and time of start of the trip': 'timestamp'
})
df_end['point_type'] = 'end'
df_end['point_order'] = 2  # Порядок точек в линии

# Объединяем
df_long = pd.concat([df_start, df_end], ignore_index=True)

# Сортируем по id_trip и порядку точек
df_long = df_long.sort_values(['id_trip', 'point_order'])

print(f"Строк после разворота: {len(df_long)}")
print(df_long.head(10))
```

```python
df_long.to_csv('taxi_trips_long_format_15376_rows.csv', index=False, encoding='utf-8')
print(f"Сохранено {len(df_long)} строк в CSV файл")
```
